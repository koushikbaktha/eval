```python
#load the model from huggingface with 50 gpu layers
llm = AutoModelForCausalLM.from_pretrained("TheBloke/Mistral-7B-Instruct-v0.1-GGUF",
model_file="mistral-7b-instruct-v0.1.Q4_K_M.gguf",
model_type = "mistral", gpu_layers = 50)
```

```python
#function to print the model output in colab in a readable manner
def colab_print(text, max_width = 120):
  words = text.split()
  line = ""
  for word in words:
    if len(line) + len(word) + 1 > max_width:
      print(line)
      line = ""
    line += word + " "
  print (line)
```

```python
#generate text
colab_print(llm('''
-----------------------------------------------------------------------------------
CONVERSATION
###agent
thank you for calling koushik rental car facility, the no#1 rental cars in united states of america, how are you today and hope everything is well for you
###customer
thank you, I am good.
###agent
please tell me about the issue that you would like to get resolved today for you
###customer
I came here to ask about the payment issue that i took on my rental car 2 days back.
-----------------------------------------------------------------------------------
CONTEXT
this is a conversation between an agent and a customer in contact center
-----------------------------------------------------------------------------------
OBJECTIVE
this conversation has to be evaluated by a supervisor if required questions were asked so that customer experience is not impacted.
use this conversation as a reference and do not provide any other outside information for the question stated below.
-----------------------------------------------------------------------------------
QUESTION
Please answer if the agent greeted the customer properly when the customer joined the call as per the CONVERSATION block and provide the appropriate agent conversation in your response?
''',
max_new_tokens = 2048, temperature = 0.9, top_k = 55, top_p = 0.93,
repetition_penalty = 1.2))
```
